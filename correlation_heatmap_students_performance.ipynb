```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance (1).csv")
df.head()
```

```python
df.info()
```

```python
score_data = df[['math score', 'reading score', 'writing score']]
score_data.head()
```

```python
correlation_matrix = score_data.corr()
correlation_matrix
```

```python
plt.figure(figsize=(8, 6))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    linewidths=0.5
)

plt.title("Correlation Heatmap of Student Exam Scores")
plt.show()
```
